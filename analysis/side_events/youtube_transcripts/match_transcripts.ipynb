```python
import pandas as pd
import numpy as np
```

### Match with event topics

```python
transcript_meta = pd.read_csv("youtube_video_metadata.csv",encoding = "utf-8")
transcript_meta
```

```python
energy_events = pd.read_csv("../extracted_events_energy.csv",encoding = "utf-8")
energy_cop28 = energy_events[energy_events["COP"]=="COP28"]

for i in transcript_meta.index:
    title = transcript_meta.loc[i,"title"]
    check_title = energy_cop28[energy_cop28["Title/Description"].str.contains(title)]
    if len(check_title)==0:
        continue
    else:
        energy_cop28.loc[check_title.index,"video_num"] = transcript_meta.loc[i,"video_num"]
energy_cop28.loc[20,"video_num"] = 68 # manual check
energy_cop28.loc[17,"video_num"] = 129 # manual check
energy_cop28.to_csv("cop28_video_match_energy.csv")
```

```python
food_events = pd.read_csv("../extracted_events_food.csv",encoding = "utf-8")
food_cop28 = food_events[food_events["COP"]=="COP28"]
for i in transcript_meta.index:
    title = transcript_meta.loc[i,"title"]
    check_title = food_cop28[food_cop28["Title/Description"].str.contains(title)]
    if len(check_title)==0:
        continue
    else:
        food_cop28.loc[check_title.index,"video_num"] = transcript_meta.loc[i,"video_num"]
food_cop28.loc[24,"video_num"] = 36 # manual match
food_cop28.to_csv("cop28_video_match_food.csv")
```
