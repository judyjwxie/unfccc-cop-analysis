### oil and gas

```python
delegates_fossil = pd.read_excel("../delegates/COP28_delegates_ff_organizations_renamed_filtered_check.xlsx")
energy_nominator_rename = pd.read_excel("nominator_energy_match.xlsx")
energy_sponsor = delegates_fossil[["nominator"]].drop_duplicates()
for i in energy_nominator_rename.index:
    energy_sponsor["nominator"] = energy_sponsor["nominator"].replace(energy_nominator_rename.loc[i,"original_name"],energy_nominator_rename.loc[i,"renamed"])
energy_sponsor = energy_sponsor.drop_duplicates()
energy_sponsor
```
