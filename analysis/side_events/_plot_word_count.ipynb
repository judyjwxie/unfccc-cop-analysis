```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
```

```python
events_lang = pd.read_csv("all_side_events_complete_lang.csv") 
events_lang
```

```python
fig, axes = plt.subplots(figsize=(6,4))
sns.histplot(data=events_lang, x="word_count",ax=axes)
axes.set_xlabel("Title and description text length (Number of words)")
fig.savefig("word_count_histogram.png",dpi=600, bbox_inches='tight')
```
