## Get admitted IGO details
https://unfccc.int/process/parties-non-party-stakeholders/non-party-stakeholders/admitted-igos/list-of-admitted-igos downloaded 13 Feb 2024

```python
url = "Admitted IGOs _ UNFCCC.html"
page = open(url, encoding="utf8")
soup = bs(page.read())
table = soup.find("table", attrs={"class":"list_table"})
```

```python
igo_table = pd.read_html(str(table),header=0)[0]
igo_table = igo_table[igo_table["Official Name"].isna()==False]
igo_table = igo_table.reset_index().drop(columns="index")
igo_table
```

```python
salutations = [" Ms. "," Mr. ", " Mme ", " Dr. "," Sr. "," Sra. ",
               " M. " ," Ind. "," Dr ","  "," Mr "," Mx. ", " H.E. "]

for i in igo_table.index:
    official_name = igo_table.loc[i,"Official Name"]
    for s in salutations:
        if s in official_name:
            split_name = official_name.split(s)
            igo_table.loc[i,"Organization"] = split_name[0]
            igo_table.loc[i,"Contact Person"] = split_name[1]

# identify short name
for k in igo_table.index:
    org_name = igo_table.loc[k,"Organization"]
    igo_table.loc[k,"Short_Org"] = org_name.split(" (")[0]

igo_table.set_index("Official Name").to_excel("org_list_igo.xlsx")
igo_table
```
