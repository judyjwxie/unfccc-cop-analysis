```python
import pandas as pd
import numpy as np
```

```python
party_data = pd.read_csv("../delegates/COP28_delegates_all.csv")
party_data = party_data.query("type=='Parties'")
party_names_list = party_data["nominator"].unique()
party_names_list
```

```python
annex_one = ["Australia","Austria","Belarus","Belgium","Bulgaria","Canada","Denmark","Estonia","Finland",
                  "France","Germany","Greece","Hungary","Iceland","Ireland","Italy","Japan","Latvia","Lithuania",
                  "Luxembourg","Netherlands (Kingdom of the)","New Zealand","Norway","Poland","Portugal","Romania",
                  "Russian Federation","Spain","Sweden","Switzerland","Türkiye","Ukraine",
                  "United Kingdom of Great Britain and Northern Ireland","United States of America","Croatia",
                  "Czechia","Liechtenstein","Monaco","Slovakia","Slovenia", "Malta","Cyprus"]
annex_two = ["Australia","Austria","Belgium","Canada","Denmark","Finland","France","Germany",
             "Greece","Iceland","Ireland","Italy","Japan","Luxembourg","Netherlands (Kingdom of the)",
             "New Zealand","Norway","Portugal","Spain","Sweden","Switzerland",
             "United Kingdom of Great Britain and Northern Ireland","United States of America"]
```

```python
party_names = pd.DataFrame(data={"party":party_names_list})
for i in party_names.index:
    if party_names.loc[i,"party"] in annex_one:
        party_names.loc[i,"annex_1"] = 1
        if party_names.loc[i,"party"] in annex_two:
            party_names.loc[i,"annex_2"] = 1
            party_names.loc[i,"type"] = "Annex II"
        else:
            party_names.loc[i,"type"] = "Annex I Only"
    else:
        party_names.loc[i,"type"] = "Non-Annex"
party_names = party_names.fillna(0)
party_names.set_index("party").to_csv("org_list_party.csv")
party_names.set_index("party").to_excel("org_list_party.xlsx")

party_names
```
