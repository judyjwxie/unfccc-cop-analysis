```python
xl = pd.ExcelFile('COP28%20FLOP%20On-site.xlsx')
sheet_names = xl.sheet_names  # see all sheet names
for sn in sheet_names:
    this_sheet = xl.parse(sn) 
    this_sheet["type"] = sn
    if sn == sheet_names[0]:
        all_sheets = this_sheet
    else:
        all_sheets = pd.concat([all_sheets,this_sheet])

# rename column names to match previous version
all_sheets = all_sheets.rename(columns={"Nominator":"nominator","Name":"name","Functional title":"func_title",
                                       "Department":"department","Organization":"organization",
                                       "Relation":"relation"})
all_sheets
```

```python
all_sheets.set_index("nominator").to_csv("COP28_delegates_all.csv")
```
